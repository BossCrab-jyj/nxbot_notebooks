# 天使过银河

![title](other_data/01.jpg)
1. 问题描述
   在河的左岸有N个天使(A)、N个恶魔(E)和一条船，天使们想通过这条船把所有天使都运到河对面去，但有以下条件限制：

   1.1天使和恶魔都会划船，但船每次最多只能运2个人；
    
   1.2在任何岸边恶魔数目都不能超过天使，否则天使会被恶魔打败。

   假定恶魔会服从任何一种过河安排，请规划出一个确保天使能够安全过河的计划。

2. 问题分析

    2.1约束条件：
       （一）： A≧E 任何时刻两岸、船上都必须满足天使人数不少于恶魔数量（A=0时除外，表示当前没有天使） 
       （二）： A+E≦2 船上人数限制在2以内

    2.2求解：
     天使与恶魔全部安全渡到对岸的解决方案
    
3. 求解过程

   3.1变量假设：
    设E=3，A=3，B=(True or False)
     L：左岸，R：右岸， 
     B：是否有船（True：有船，False：无船） 
     
   3.2状态表示
    定义：用列表 [AS，ES，BS] 表示起始岸状态，其中：
    
    0≦AS≦3,0≦ES≦3，BS∈｛True,False｝
    
    如：[3，1，True] 表示岸边有三个天使，一个恶魔，并且船也在起始岸。
    
    目标：从 [3，3，True] 到 [0，0，False] 的状态转换
    
    状态空间：4 * 4 * 2=32 种状态，
    
    其中：
    12种不合理状态：如 [1，2，False] 说明岸边有1个天使，2个恶魔；
    
    4种不可能状态：如 [3，3，False] 说明所有A和E都在左岸，而船在右岸 ,以及其它三种状态[0,0,True]  [1,0,True]  [0,1,True]。
    

## 合理状态

```python
def is_legal(new_Angle,new_Evil,boatAdr):
    
    # 天使和恶魔都是3个，如果没在这个范围内就返回False
    if new_Evil < 0 or new_Evil > 3 or new_Angle < 0 or new_Angle > 3:
        return False
    else:
        # 查看当前状态是否存在于状态列表中
        global temp_state
        for i in range(0,len(temp_state)):
            # 如果已经存在与状态列表中就返回False
            if temp_state[i][0]==new_Angle and temp_state[i][1]==new_Evil and temp_state[i][2]!=boatAdr:
                return False
    # 如果（天使数量等于0，或者天使数量大于等于恶魔），（如果对岸的天使数量等于3或者对岸的天使数量大于恶魔数量）则返回 True
    if (new_Angle == 0 or new_Evil <= new_Angle)  and ((new_Angle == 3) or (3 - new_Evil) <= (3 - new_Angle)):
        return True
    else:
        return False
```

## 开始游戏
通过设定的规则改变岸边和对岸的状态直到岸边状态为[0，0，0]则游戏结束。

```python
# 用于保存移动的步骤
result = []
# 穿上乘客数量
boat_size=2
# 设置一个列表，用于记录起始岸的状态信息。
temp_state = [[3,3,True]]

# 游戏主函数
def AngleEvil(num_Angle,num_Evil,boatAdr):
    # 达到目标状态就退出
    if num_Angle ==0 and num_Evil==0 and boatAdr==False:
        result.append(temp_state[:])
        return
    # 如果有船代表要过河
    if boatAdr == True:
        i = boat_size
        # 直到船坐满为止
        while i>0:
            # 乘客依次上船
            for j in range(0,i+1):
                # 遍历所有可能组合的状态
                # 当初始状态为 [3,3,True], i=1，j=0时,[num_Angle,num_Evil,boatAdr] = [3-0,3-(1-0),True] = [3,2,True]
                # 当初始状态为 [3,3,True], i=1, j=1时,[num_Angle,num_Evil,boatAdr] = [3-1,3-(1-1),True] = [2,3,True]
                # 当初始状态为 [3,3,True], i=2，j=0时,[num_Angle,num_Evil,boatAdr] = [3-0,3-(2-0),True] = [3,1,True]
                # 当初始状态为 [3,3,True], i=2, j=1时,[num_Angle,num_Evil,boatAdr] = [3-1,3-(2-1),True] = [2,2,True]
                # 当初始状态为 [3,3,True], i=2, j=2时,[num_Angle,num_Evil,boatAdr] = [3-2,3-(2-2),True] = [1,3,True]
                new_Angle = num_Angle - j
                new_Evil = num_Evil-(i-j)
                
                # 如果当前状态合法将这个状态记录下来
                if is_legal(new_Angle, new_Evil, boatAdr):
                    # 如果移动了，那么说明当前boatAdr就为 False了
                    temp_state.append([new_Angle,new_Evil, not boatAdr])
                    # 然后将剩下的天使与恶魔的数量再次进行进行以上操作，并且将满足条件的状态保存下来，直到岸边状态为[0,0,False]就停止
                    AngleEvil(new_Angle, new_Evil, not boatAdr)
                    # 使用先进后出的原则，顺着初始状态往下走，直到无路可走就退回去找另外一条路径，直到到达目标状态
                    temp_state.pop()
            i=i-1

    # 如果没有船则代表回到起始岸
    else:
        i = boat_size
        while i>0:
            for j in range(0, i + 1):
                new_Angle = num_Angle + j
                new_Evil = num_Evil + (i - j)
                if is_legal(new_Angle, new_Evil, boatAdr):
                    temp_state.append([new_Angle, new_Evil, not boatAdr])
                    AngleEvil(new_Angle, new_Evil, not boatAdr)
                    temp_state.pop()
            i=i-1
            
AngleEvil(3,3,True)
```

## 打印过河的步骤与状态空间

```python
def print_state(result):
    i = 1
    print('一共有%d个解'%(len(result)))
    # 遍历所有结果
    for states in result:
        # 初始化起始岸与目标岸的状态信息
        start_state=[3,3,1]
        print('第%d个解：'%(i))
        print('起始岸状态空间为：',start_state,'\n')
        target_state= [0,0,0]
        # 每一种可以到达目标岸的状态列表
        for state in range(0,len(states)-1):
            # 如果起始岸有船代表当前乘客可以移动到目标岸
            if states[state][2]==True:
                # 求出每次移动了多少个天使与恶魔
                # states[state][0]  :可以到达目标状态列表中的当前状态的天使数量
                # states[state+1][0]:可以到达目标状态列表中的下一次状态的天使数量
                cross_angel = states[state][0]-states[state+1][0]
                cross_evil  = states[state][1]-states[state+1][1]
                # 移动后的起始岸状态空间为[Angle, Evil, False],也就可以计为[Angle, Evil, 0]
                start_state = [start_state[0]-cross_angel, start_state[1]-cross_evil, 0]
                # 移动后的目标岸状态空间
                target_state[0] = 3-start_state[0]
                target_state[1] = 3-start_state[1]
                target_state[2] = 1-start_state[2]
                print("%s个天使和%s个恶魔移动到目标点"%(cross_angel,cross_evil))
                print('起始岸状态空间为：', start_state)
                print('目标岸状态空间为：', target_state, '\n')
            # 如果目标岸有船代表当前乘客可以返回到起始岸
            else:
                back_angel = states[state+1][0]-states[state][0]
                back_evil  = states[state+1][1]-states[state][1]
                # 移动后的起始岸状态空间为[Angle, Evil, True],也就可以计为[Angle, Evil, 1]
                start_state = [start_state[0]+back_angel, start_state[1]+back_evil, 1]
                target_state[0] = 3-start_state[0]
                target_state[1] = 3-start_state[1]
                target_state[2] = 1-start_state[2]
                print("%s个天使和%s个恶魔回到起始点"%(back_angel,back_evil))
                print('起始岸状态空间为：', start_state)
                print('目标岸状态空间为：', target_state, '\n')
        i+=1
        print('\n')
        
print_state(result)
```
