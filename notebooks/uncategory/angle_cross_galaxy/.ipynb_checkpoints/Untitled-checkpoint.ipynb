# 天使过银河
1. 问题重述
   在河的左岸有N个天使(A)、N个恶魔(E)和一条船，天使们想通过这条船把所有天使都运到河对面去，但有以下条件限制：

    （1）天使和恶魔都会划船，但船每次最多只能运K个人；
    （2）在任何岸边恶魔数目都不能超过天使，否则天使会被恶魔打败。

   假定恶魔会服从任何一种过河安排，请规划出一个确保天使能够安全过河的计划。

2. 问题分析

    2.1约束条件：
        ① A≧E 任何时刻两岸、船上都必须满足天使人数不少于恶魔数量（A=0时除外，既没有天使） 
        ② A+E≦K 船上人数限制在K以内

    2.2求解：
     天使与恶魔全部安全渡到对岸的解决方案
    
3. 求解过程

   3.1变量假设：
    设N=3，K=2（三个A和三个E，每次渡河二人以下）
     L：左岸，R：右岸， 
     B：是否有船（0：无船，1：有船） 
     
   3.2状态表示
    定义：用三元组（AL，EL，BL）表示左岸状态，其中：
    0≦AL,EL≦3，BL∈｛0,1｝
    如：（0，3，1）表示左岸有三个恶魔，船在左岸。
    从（3，3，1）到（0，0，0）的状态转换
    状态空间：32 种状态，其中：
    12种不合理状态：如（1，0，1）说明右岸有2个A，3个E；
    4种不可能状态：如（3，3，0）说明所有A和E都在左岸，而船在右岸

## 合理状态

```python
def is_legal(temp_state, num_Angle,num_Evil,boatAdr):
    if num_Evil < 0 or num_Evil > 3 or num_Angle < 0 or num_Angle > 3:
        return False
    else:
        for i in range(0,len(temp_state)):
            if temp_state[i][0] ==num_Angle and temp_state[i][1]==num_Evil and temp_state[i][2]==boatAdr:
                return False
    if (num_Angle == 0 or num_Evil <= num_Angle) and ((3 - num_Angle) == 0 or (3 - num_Evil) <= (3 - num_Angle)):
        return True
    else:
        return False
```

```python
result = []
boat_size=2
temp_state = [[3,3,True]]
# count = 0
def AngleEvil(num_Angle,num_Evil,boatAdr):
    if num_Angle ==0 and num_Evil==0 and boatAdr==False:
        result.append(temp_state[:])
        return
    
    if boatAdr == True:
        i = boat_size
        while i>0:
            for j in range(0,i+1):
                if is_legal(temp_state, num_Angle - j, num_Evil-(i-j), not boatAdr):
                    temp_state.append([num_Angle - j,num_Evil-(i-j),not boatAdr])
                    AngleEvil(num_Angle - j, num_Evil-(i-j), not boatAdr)
                    temp_state.pop()
            i=i-1
    else:
        i = boat_size
        while i>0:
            for j in range(0, i + 1):
                if is_legal(temp_state, num_Angle + j, num_Evil + (i - j), not boatAdr):
                    temp_state.append([num_Angle + j, num_Evil + (i - j), not boatAdr])
                    AngleEvil(num_Angle + j, num_Evil + (i - j), not boatAdr)
                    temp_state.pop()
            i=i-1
    
```

```python
def print_state(result):
    for i in result:
        for j in range(0,len(i)-1):
            if i[j][2]==True:
                cross_evil  = i[j][1]-i[j+1][1]
                cross_angel = i[j][0]-i[j+1][0]
                print("%s个恶魔和%s个天使过河"%(cross_evil,cross_angel))
                print('对岸状态空间为：', cross_evil, cross_angel, 1)
                print('岸边状态空间为：', 3-cross_evil, 3-cross_angel, 0, '\n')
            else:
                back_evil  = i[j+1][1]-i[j][1]
                back_angel = i[j+1][0]-i[j][0]
                print("%s个恶魔和%s个天使回来" % (back_evil,back_angel))
                print('对岸状态空间为：', back_evil, back_angel, 0)
                print('岸边状态空间为：', 3-back_evil, 3-back_angel, 1, '\n')
```

```python
def sum_numbers(num):
 
    if num == 1:
        return 1
    # 数字的累加 num + (1.....+ num-1)
    # 假设sum_numbers能够正确处理1....num-1
    temp = sum_numbers(num-1)
    return temp + num
 

gl_num = int(input())
print(sum_numbers(gl_num))
```

```python
class Solution:
    boatnum=2 #船装人的个数
    answer =[] #返回的结果
    temp_state=[[3,3,True]] #无意义的值 中间件
    '''将代码翻译成通俗易懂的语言'''
    def print_state(self):
        for i in self.answer:
            for j in range(0,len(i)-1):
                if i[j][2]==True:
                    print("%s个恶魔和%s个天使过河"%(i[j][1]-i[j+1][1],i[j][0]-i[j+1][0]))
                    print('对岸状态空间为：', i[j][1]-i[j+1][1], i[j][0]-i[j+1][0], 1)
                    print('岸边状态空间为：', 3-(i[j][1]-i[j+1][1]), 3-(i[j][0]-i[j+1][0]), 0, '\n')
                else:
                    print("%s个恶魔和%s个天使回来" % (i[j+1][1]-i[j][1],i[j+1][0]-i[j][0]))
                    print('对岸状态空间为：', i[j+1][1]-i[j][1], i[j+1][0]-i[j][0], 0)
                    print('岸边状态空间为：', 3-(i[j+1][1]-i[j][1]), 3-(i[j+1][0]-i[j][0]), 1, '\n')
                    

    '''判断是否合法'''
    def is_legal(self,num_Angle,num_Evil,boatAdr):
        if num_Evil < 0 or num_Evil > 3 or num_Angle < 0 or num_Angle > 3:
            return False
        else:
            for i in range(0,len(self.temp_state)):
                if self.temp_state[i][0] ==num_Angle and self.temp_state[i][1]==num_Evil and self.temp_state[i][2]==boatAdr:
                    return False
        if (num_Angle == 0 or num_Evil <= num_Angle) and ((3 - num_Angle) == 0 or (3 - num_Evil) <= (3 - num_Angle)):
            return True
        else:
            return False

    '''深度搜索总代码'''
    def AngleEvil(self,num_Angle,num_Evil,boatAdr) :
        if num_Angle ==0 and num_Evil==0 and boatAdr==False:
            self.answer.append(self.temp_state[:])
            return

        if boatAdr == True:
            i = self.boatnum;
            while i>0:
                for j in range(0,i+1):
                    if self.is_legal(num_Angle - j, num_Evil-(i-j), not boatAdr):
                        self.temp_state.append([num_Angle - j,num_Evil-(i-j),not boatAdr])
                        self.AngleEvil(num_Angle - j, num_Evil-(i-j), not boatAdr)
                        self.temp_state.pop()
                i=i-1
        else:
            i = self.boatnum;
            while i > 0:
                for j in range(0, i + 1):
                    if self.is_legal(num_Angle + j, num_Evil + (i - j), not boatAdr):
                        self.temp_state.append([num_Angle + j, num_Evil + (i - j), not boatAdr])
                        self.AngleEvil(num_Angle + j, num_Evil + (i - j), not boatAdr)
                        self.temp_state.pop()
                i = i - 1
    def __init__(self):
        self.AngleEvil(3,3,True)
        self.print_state()
if __name__ == "__main__":
    Solution()
```
